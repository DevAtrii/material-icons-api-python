```python
from py_utils.common_utils.file_utils import *

data = open_file_as_json('data.json')
```

```python
# data = data['icons']
# print(data)
tags_w_icons = {}
# categories_w_icons = {}
for i,icon in enumerate(data):
    tags = icon.get('tags',[])
    # categories = icon.get('categories',[])
    icon_name = icon['name']
    for tag in tags:
        if tag not in tags_w_icons:
            tag_dict = {
                tag:[icon_name]
            }
            tags_w_icons.update(tag_dict)
            continue
        tags_w_icons[tag].append(icon_name)
    # for category in categories:
    #     if category not in categories_w_icons:
    #         cat_dict = {
    #             category:[icon_name]
    #         }
    #         categories_w_icons.update(cat_dict)
    #         continue
    #     categories_w_icons[category].append(icon_name)
print(tags_w_icons)
```

```python
icons_db_w_tag = {

}
for i,tag  in enumerate(tags_w_icons):
    tag_icons = tags_w_icons[tag]
    for icon in tag_icons:
        icon_folder = f'svg/{icon}/'
        files = get_all_files_in_dir(icon_folder)
        varients = []
        for file in files:
            varients.append(file.replace('.svg',''))
        icon_dict = {
            'name':icon,
            'varients':varients
        }
        if tag not in icons_db_w_tag:
            tag_dict = {
                tag:[icon_dict]
            }
            icons_db_w_tag.update(tag_dict)
            continue
        icons_db_w_tag[tag].append(icon_dict)
```

```python
save_json_file(path='icons_db.json',content=icons_db_w_tag,increment=False)
```
